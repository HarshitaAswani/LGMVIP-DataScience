```python
# importing the necessary libraries
import cv2
import matplotlib.pyplot as plt
```

```python
# reading the image
image = cv2.imread('C:\\Users\\ASUS\\Downloads\\tajmahal.jpg')
```

```python
# viewing the image
plt.title('Original taj')
plt.imshow(image)
plt.show()
```

```python
# converting original image to greyscale
greyimage=cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
# viewing the image
plt.title('Grey taj')
plt.imshow(greyimage, cmap='gray')
plt.show()
```

```python
# inverting the new greyscale image
invertedgreyimage= 255-greyimage
# viewing the image
plt.title('Inverted Grey taj')
plt.imshow(invertedgreyimage, cmap='gray')
plt.show()
```

```python
# blur the image
blurimage=cv2.GaussianBlur(invertedgreyimage, (21, 21), 0)
# viewing the image
plt.title('Blur taj')
plt.imshow(blurimage, cmap='gray')
plt.show()
```

```python
# inverting the new blur image
invertedblurimage= 255-blurimage
# viewing the image
plt.title('Inverted Blur taj')
plt.imshow(invertedblurimage, cmap='gray')
plt.show()
```

```python
# converting the image into pencil sketch
pencilsketch=cv2.divide(greyimage, invertedblurimage, scale=256.0)
# viewing the image
plt.title('Sketch of taj')
plt.imshow(pencilsketch, cmap='gray')
plt.show()
```
